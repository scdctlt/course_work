## 导入相关工具包

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm
import os
import tensorflow.keras
from tensorflow.keras.models import *
from sklearn.utils import shuffle
from tensorflow.keras.layers import *
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications import resnet50
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, TensorBoard, ModelCheckpoint
from sklearn.metrics import classification_report,confusion_matrix
import ipywidgets as widgets
import io
from PIL import Image
from IPython.display import display,clear_output
from warnings import filterwarnings
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False 
colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
colors_red = ["#331313", "#582626", '#9E1717', '#D35151', '#E9B4B4']
colors_green = ['#01411C','#4B6F44','#4F7942','#74C365','#D0F0C0']
```

## 设置固定参数

```python
labels = ['glioma','notumor','meningioma','pituitary']
X_train = []
y_train = []
image_size = 128
epochs=20
batch_size=16
loss='categorical_crossentropy'
optimizer = 'Adam'
metrics= ['accuracy']
```

## 读取数据进行相关处理

### 读取及预处理

```python
for i in labels:
    folderPath = os.path.join('data/Training',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size, image_size))
        X_train.append(img)
        y_train.append(i)
        
for i in labels:
    folderPath = os.path.join('data/Testing',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size,image_size))
        X_train.append(img)
        y_train.append(i)
X_train = np.array(X_train)
y_train = np.array(y_train)
```

### 查看部分样本

```python
k=0
fig, ax = plt.subplots(1,4,figsize=(20,20))
for i in labels:
    j=0
    while True :
        if y_train[j]==i:
            ax[k].imshow(X_train[j])
            ax[k].set_title(y_train[j])
            ax[k].axis('off')
            k+=1
            break
        j+=1
```

### 数据集划分

```python
X_train, y_train = shuffle(X_train,y_train, random_state=101)
X_train.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train, test_size=0.2,random_state=101)
```

### 查看各类型占比

```python
from palettable.colorbrewer.qualitative import Pastel1_7
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(10,5))
axs[0].set_title('各类别占比') 
axs[0].pie(np.unique(y_train,return_counts=True)[1],
         labels=['pituitary','notumor', 'meningioma', 'glioma'],
         colors=Pastel1_7.hex_colors,wedgeprops=dict(width=0.3, edgecolor='w'),
         autopct='%.1f%%') 
axs[1].set_title('训练集测试集比例')
axs[1].pie([len(y_train), 
               len(y_test)],
               labels=['Train','Test'],
               autopct='%.1f%%', explode=(0.05,0),
               startangle=30);

fig.show()
```

```python
y_train_new = []
for i in y_train:
    y_train_new.append(labels.index(i))
y_train = y_train_new
y_train = tf.keras.utils.to_categorical(y_train)


y_test_new = []
for i in y_test:
    y_test_new.append(labels.index(i))
y_test = y_test_new
y_test = tf.keras.utils.to_categorical(y_test)
```

## 模型部分

### EfficientNetB0

#### 模型导入

```python
effnet = EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size,image_size,3))
```

#### 搭建模型

```python
# effnet_b0 = effnet.output
# effnet_b0 = tf.keras.layers.GlobalAveragePooling2D()(effnet_b0)
# effnet_b0 = tf.keras.layers.Dropout(rate=0.5)(effnet_b0)
# effnet_b0 = tf.keras.layers.Dense(4,activation='softmax')(effnet_b0)
# effnet_b0 = tf.keras.models.Model(inputs=effnet.input, outputs = effnet_b0)
effnetb0_model = Sequential()
effnetb0_model.add(Input(shape=(image_size,image_size,3)))
effnetb0_model.add(effnet)
effnetb0_model.add(Flatten())
effnetb0_model.add(Dropout(0.3))
effnetb0_model.add(Dense(128, activation='relu'))
effnetb0_model.add(Dropout(0.2))
effnetb0_model.add(Dense(len(labels), activation='softmax'))
```

```python
effnetb0_model.summary()
```

#### 查看模型结构图

```python
tf.keras.utils.plot_model(effnetb0_model, show_shapes=True)
```

#### 设置tensorboard和自动调整学习率

```python
effnetb0_model.compile(loss=loss,optimizer = optimizer, metrics=metrics)
tensorboard = TensorBoard(log_dir = 'logs')
checkpoint = ModelCheckpoint("effnetb0_model.h5",monitor="val_accuracy",save_best_only=True,mode="auto",verbose=1)
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.3, patience = 2, min_delta = 0.001,
                              mode='auto',verbose=1)
```

#### 开始训练

```python
%time
history_effnet_b0 = effnetb0_model.fit(X_train,y_train,validation_split=0.1, epochs =epochs, verbose=1, batch_size=batch_size,
                   callbacks=[tensorboard,checkpoint,reduce_lr])
```

#### 绘制Accuracy/Loss随epochs变化情况图

```python
filterwarnings('ignore')
history_effnet_b0

epochs = [i for i in range(epochs)]
fig, ax = plt.subplots(1,2,figsize=(14,7))
train_acc = history_effnet_b0.history['accuracy']
train_loss = history_effnet_b0.history['loss']
val_acc = history_effnet_b0.history['val_accuracy']
val_loss = history_effnet_b0.history['val_loss']

fig.text(s='EfficientNetB0: Accuracy/Loss',size=18,fontweight='bold',
             fontname='monospace',color=colors_dark[1],y=1,x=0.28,alpha=0.8)

sns.despine()
ax[0].plot(epochs, train_acc, marker='o',markerfacecolor=colors_green[2],color=colors_green[3],
           label = 'Training Accuracy')
ax[0].plot(epochs, val_acc, marker='o',markerfacecolor=colors_red[2],color=colors_red[3],
           label = 'Validation Accuracy')
ax[0].legend(frameon=False)
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Accuracy')

sns.despine()
ax[1].plot(epochs, train_loss, marker='o',markerfacecolor=colors_green[2],color=colors_green[3],
           label ='Training Loss')
ax[1].plot(epochs, val_loss, marker='o',markerfacecolor=colors_red[2],color=colors_red[3],
           label = 'Validation Loss')
ax[1].legend(frameon=False)
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Training & Validation Loss')

fig.show()
```

#### 模型性能

```python
pred = effnetb0_model.predict(X_test)
pred = np.argmax(pred,axis=1)
y_test_new = np.argmax(y_test,axis=1)
print(classification_report(y_test_new,pred))
```

#### 绘制混淆矩阵

```python
fig,ax=plt.subplots(1,1,figsize=(14,7))
sns.heatmap(confusion_matrix(y_test_new,pred),ax=ax,xticklabels=labels,yticklabels=labels,annot=True,
           cmap=colors_green[::-1],alpha=0.7,linewidths=2,linecolor=colors_dark[3],fmt='.20g')
fig.text(s='Heatmap of the Confusion Matrix',size=18,fontweight='bold',
             fontname='monospace',color=colors_dark[1],y=0.92,x=0.28,alpha=0.8)

plt.show()
```

### Renet50

#### 模型导入

```python
resnet = resnet50.ResNet50(weights='imagenet', input_shape=(image_size,image_size,3), include_top=False)

for layer in resnet.layers[:171]:
    layer.trainable = False

for layer in resnet.layers[171:]:
    layer.trainable = True
```

#### 模型搭建

```python
resnet_model = Sequential()
resnet_model.add(Input(shape=(image_size,image_size,3)))
resnet_model.add(resnet)
resnet_model.add(Flatten())
resnet_model.add(Dropout(0.3))
resnet_model.add(Dense(128, activation='relu'))
resnet_model.add(Dropout(0.2))
resnet_model.add(Dense(4, activation='softmax')) 
resnet_model.summary()
```

#### 查看模型结构图

```python
tf.keras.utils.plot_model(resnet_model, show_shapes=True)
```

#### 设置参数和tensorboard

```python
resnet_model.compile(loss=loss,optimizer = optimizer, metrics=metrics)
tensorboard = TensorBoard(log_dir = 'logs')
checkpoint = ModelCheckpoint("resnet.h5",monitor="val_accuracy",save_best_only=True,mode="auto",verbose=1)
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.3, patience = 2, min_delta = 0.001,
                              mode='auto',verbose=1)
```

#### 开始训练resnet_model

```python
%time
epochs=20
batch_size=16
history_resnet = resnet_model.fit(X_train,y_train,validation_split=0.1, epochs =epochs, verbose=1, batch_size=batch_size,
                   callbacks=[tensorboard,checkpoint,reduce_lr])
```

#### 绘制Accuracy/Loss随epochs变化情况图

```python
filterwarnings('ignore')
history_resnet

epochs = [i for i in range(epochs)]
fig, ax = plt.subplots(1,2,figsize=(14,7))
train_acc = history_resnet.history['accuracy']
train_loss = history_resnet.history['loss']
val_acc = history_resnet.history['val_accuracy']
val_loss = history_resnet.history['val_loss']

fig.text(s='ResNet50: Accuracy/Loss',size=18,fontweight='bold',
             fontname='monospace',color=colors_dark[1],y=1,x=0.28,alpha=0.8)

sns.despine()
ax[0].plot(epochs, train_acc, marker='o',markerfacecolor=colors_green[2],color=colors_green[3],
           label = 'Training Accuracy')
ax[0].plot(epochs, val_acc, marker='o',markerfacecolor=colors_red[2],color=colors_red[3],
           label = 'Validation Accuracy')
ax[0].legend(frameon=False)
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Accuracy')

sns.despine()
ax[1].plot(epochs, train_loss, marker='o',markerfacecolor=colors_green[2],color=colors_green[3],
           label ='Training Loss')
ax[1].plot(epochs, val_loss, marker='o',markerfacecolor=colors_red[2],color=colors_red[3],
           label = 'Validation Loss')
ax[1].legend(frameon=False)
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Training & Validation Loss')

fig.show()
```

#### 预测值

```python
pred = resnet_model.predict(X_test)
pred = np.argmax(pred,axis=1)
y_test_new = np.argmax(y_test,axis=1)
print(classification_report(y_test_new,pred))
```

#### 绘制混淆矩阵

```python
fig,ax=plt.subplots(1,1,figsize=(14,7))
sns.heatmap(confusion_matrix(y_test_new,pred),ax=ax,xticklabels=labels,yticklabels=labels,annot=True,
           cmap=colors_green[::-1],alpha=0.7,linewidths=2,linecolor=colors_dark[3],fmt='.20g')
fig.text(s='Heatmap of the Confusion Matrix',size=18,fontweight='bold',
             fontname='monospace',color=colors_dark[1],y=0.92,x=0.28,alpha=0.8)

plt.show()
```
